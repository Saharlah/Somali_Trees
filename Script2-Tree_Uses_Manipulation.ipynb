```python
import pandas as pd
import os
```

```python
dataset = pd.read_csv("/Users/saharlahershi/Desktop/Somali_Trees_Data/Somali_Trees_Cleaned/Tree_Uses.csv")
dataset
```

```python
dataset_modified = dataset.copy()
```

```python
value_variables = list(dataset_modified.columns)[1:]
value_variables
```

```python
dataset_melted = dataset_modified.melt(id_vars = "Botanical_Name",value_vars= value_variables, var_name="Tree_Uses", value_name= "Usefulness")
dataset_melted
```

```python
dataset_melted.isna().sum()
```

```python
dataset_final = dataset_melted.dropna(subset= ["Usefulness"])
dataset_final
```

```python
dataset_final.to_csv("/Users/saharlahershi/Desktop/Somali_Trees_Data/Somali_Trees_Cleaned/Tree_Uses_Final.csv",index=False)
```
