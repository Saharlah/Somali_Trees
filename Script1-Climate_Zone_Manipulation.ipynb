```python
dataset_melted = dataset_modified.melt(id_vars="Botanical_Name",value_vars=["A","B","C","D","E"],var_name="Climate_Zone",value_name="Status")
dataset_melted
```

```python
dataset_nulls_dropped = dataset_melted.dropna(subset= ["Status"])
dataset_nulls_dropped
```
